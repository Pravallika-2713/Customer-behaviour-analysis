```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
df = pd.read_csv("superstore_final_dataset (1).csv", encoding='latin1')   # Updated with the correct file name and encoding
print(df.head())
print(df.info())

# Step 3: Data Cleaning
# Handle missing values
df = df.dropna()
# Remove duplicates
df = df.drop_duplicates()

# Convert date columns if present
if 'Order_Date' in df.columns:
    df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)
if 'Ship_Date' in df.columns:
    df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], dayfirst=True)

# Step 4: Sales Trends
monthly_sales = df.groupby(df['Order_Date'].dt.to_period("M"))['Sales'].sum()
monthly_sales.plot(kind='line', figsize=(10,5), title="Monthly Sales Trend")
plt.show()

# Step 5: Product Performance
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title("Sales by Product Category")
plt.show()

subcategory_sales = df.groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=subcategory_sales.index, y=subcategory_sales.values)
plt.title("Top 10 Sub-Categories by Sales")
plt.xticks(rotation=45)
plt.show()

# Step 6: Regional Performance
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title("Sales by Region")
plt.show()

# Step 7: Segment Analysis
segment_sales = df.groupby('Segment')['Sales'].sum()
sns.barplot(x=segment_sales.index, y=segment_sales.values)
plt.title("Sales by Customer Segment")
plt.show()

# Step 8: Recommendations
recommendations = [
    "Focus marketing on top-performing regions to maximize revenue.",
    "Improve supply chain efficiency in low-performing regions.",
    "Upsell high-margin products in categories with strong demand.",
    "Target corporate customers with bulk purchase discounts.",
    "Plan seasonal promotions based on monthly sales peaks."
]
for rec in recommendations:
    print("-", rec)
```
