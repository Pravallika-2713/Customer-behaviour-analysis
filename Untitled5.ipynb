The dataset (`ecommerce_customer_data_custom_ratios.csv.zip`) has been uploaded. The following cell will extract the CSV from the zip file and load it into a pandas DataFrame.

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import zipfile
import io

# Step 2: Load dataset
# Assuming the zip file 'ecommerce_customer_data_custom_ratios.csv.zip' is in the content directory.
zip_file_path = '/content/ecommerce_customer_data_custom_ratios.csv.zip'
csv_file_name_in_zip = 'ecommerce_customer_data_custom_ratios.csv' # Adjust if the CSV name inside differs

# Unzip the file and load the CSV
with zipfile.ZipFile(zip_file_path, 'r') as z:
    with z.open(csv_file_name_in_zip) as f:
        df = pd.read_csv(f)

print("Successfully loaded data from 'ecommerce_customer_data_custom_ratios.csv' within the zip file.")
print(df.head())
print(df.info())

# Step 3: Data Cleaning
# Handle missing values
df = df.dropna()   # or df.fillna(0)
# Remove duplicates
df = df.drop_duplicates()

# Convert date columns if present
if 'TransactionDate' in df.columns:
    df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
else:
    print("Warning: 'TransactionDate' column not found. Skipping date conversion.")

# Step 4: Feature Engineering (RFM)
# Ensure required columns exist for RFM
required_rfm_cols = ['CustomerID', 'TransactionDate', 'TransactionID', 'Amount']
if not all(col in df.columns for col in required_rfm_cols):
    missing_cols = [col for col in required_rfm_cols if col not in df.columns]
    raise ValueError(f"Missing required columns for RFM analysis: {missing_cols}. Please ensure your dataset contains these columns.")

# Recency: days since last purchase
recency = df.groupby('CustomerID')['TransactionDate'].max().reset_index()
recency['Recency'] = (df['TransactionDate'].max() - recency['TransactionDate']).dt.days

# Frequency: number of purchases
frequency = df.groupby('CustomerID')['TransactionID'].count().reset_index()
frequency.columns = ['CustomerID','Frequency']

# Monetary: total spend
monetary = df.groupby('CustomerID')['Amount'].sum().reset_index()
monetary.columns = ['CustomerID','Monetary']

# Merge RFM
rfm = recency.merge(frequency,on='CustomerID').merge(monetary,on='CustomerID')
print(rfm.head())

# Step 5: Clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto') # Added n_init='auto' to silence warning
rfm['Cluster'] = kmeans.fit_predict(rfm[['Recency','Frequency','Monetary']])

# Step 6: Visualization
sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Cluster', palette='Set2')
plt.title("Customer Segments")
plt.show()

# Step 7: Segment Profiling
segment_summary = rfm.groupby('Cluster').mean()
print(segment_summary)

# Step 8: Recommendations
recommendations = [
    "Reward loyal customers with exclusive offers.",
    "Re‑engage inactive customers with targeted campaigns.",
    "Upsell premium products to high‑value segments.",
    "Offer discounts to at‑risk customers with declining frequency.",
    "Personalize onboarding for new customers to improve retention."
]
for rec in recommendations:
    print("-", rec)
```
